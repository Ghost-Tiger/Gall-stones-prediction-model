```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gall_stones = pd.read_csv(r"C:\Users\GSM\Documents\ML\dataset-uci.csv")
```

###Gallstone Status: 0 (No), 1 (Yes)
- Gender: 0 (Male), 1 (Female)
- Comorbidity: 0 (No comorbidities present),1 (One comorbid condition), 2 (Two comorbid conditions), 3 (Three or more comorbid conditions)
- Coronary Artery Disease: 0 (No), 1 (Yes) 
- Hypothyroidism: 0 (No), 1 (Yes)
- Hyperlipidemia: 0 (No), 1 (Yes)
- Diabetes Mellitus: 0 (No), 1 (Yes)
- Hepatic Fat Accumulation (HFA): 0 (No fat accumulation),1 (Grade 1 (mild)), 2 (Grade 2 (moderate)), 3 (Grade 3 (severe)), 4 (Grade 4 (very severe))

```python
#A model for predicting the gall stone status of patients. y represent the status 1=Yes and 0=No
#x represents the features i.e risk factors that i'll be using to train the model
y= gall_stones['Gallstone Status']
x= gall_stones.drop('Gallstone Status', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
#After scaling the data to make it easier to train, i'll split it
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y, test_size=0.3, random_state=42)
```

```python
#training the linear regression model
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

#evaluation of the model

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'The Accuracy of the model was {accuracy:.2f}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
